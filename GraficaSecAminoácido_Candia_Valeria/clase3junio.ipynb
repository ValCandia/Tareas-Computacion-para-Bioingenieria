```python
from Bio import SeqIO

# Diccionario para mapear códigos de una letra a nombres completos
aa_names = {
    'A': 'Alanina',
    'C': 'Cisteína',
    'D': 'Ácido aspártico',
    'E': 'Ácido glutámico',
    'F': 'Fenilalanina',
    'G': 'Glicina',
    'H': 'Histidina',
    'I': 'Isoleucina',
    'K': 'Lisina',
    'L': 'Leucina',
    'M': 'Metionina',
    'N': 'Asparagina',
    'P': 'Prolina',
    'Q': 'Glutamina',
    'R': 'Arginina',
    'S': 'Serina',
    'T': 'Treonina',
    'V': 'Valina',
    'W': 'Triptófano',
    'Y': 'Tirosina',
    '*': 'STOP'  # Codón de parada
}

# Cargar el archivo FASTQ
fq_dict = SeqIO.index("SRR494102.fastq", "fastq")

# Obtener la secuencia de ADN
adn_sequence = fq_dict["SRR494102.20000"].seq

# Convertir a ARN
arn_sequence = adn_sequence.transcribe()

# Convertir a ARN y traducir a proteína
protein_sequence = adn_sequence.transcribe().translate()

# Convertir la secuencia de proteína a nombres completos
protein_names = [aa_names[aa] for aa in protein_sequence]

print("Secuencia de ADN:", adn_sequence)
print("Secuencia de ARN:", arn_sequence)
print("Secuencia de aminoácidos (abreviada):", protein_sequence)
print("\nSecuencia de aminoácidos (nombres completos):")
print(" - ".join(protein_names))
```

```python
from Bio import SeqIO
from rdkit import Chem
from rdkit.Chem import AllChem
import py3Dmol

# Diccionario para mapear códigos de una letra a nombres completos
aa_names = {
    'A': 'Alanina',
    'C': 'Cisteína',
    'D': 'Ácido aspártico',
    'E': 'Ácido glutámico',
    'F': 'Fenilalanina',
    'G': 'Glicina',
    'H': 'Histidina',
    'I': 'Isoleucina',
    'K': 'Lisina',
    'L': 'Leucina',
    'M': 'Metionina',
    'N': 'Asparagina',
    'P': 'Prolina',
    'Q': 'Glutamina',
    'R': 'Arginina',
    'S': 'Serina',
    'T': 'Treonina',
    'V': 'Valina',
    'W': 'Triptófano',
    'Y': 'Tirosina',
    '*': 'STOP'
}

# Cargar el archivo FASTQ y obtener la secuencia de proteína
fq_dict = SeqIO.index("SRR494102.fastq", "fastq")
adn_sequence = fq_dict["SRR494102.20000"].seq
protein_sequence = adn_sequence.transcribe().translate()

# Convertir la secuencia de proteína a formato de una letra (ej: "SNHHDHPFTNHH")
protein_seq_str = str(protein_sequence)

# ---- Generar la estructura 3D con RDKit ----
# Crear un péptido a partir de la secuencia
peptide = Chem.rdmolfiles.MolFromSequence(protein_seq_str)

# Añadir hidrógenos y optimizar la geometría (MMFF94)
peptide = Chem.AddHs(peptide)
AllChem.EmbedMolecule(peptide, AllChem.ETKDG())
AllChem.MMFFOptimizeMolecule(peptide)

# Guardar el péptido en formato PDB (opcional)
Chem.rdmolfiles.MolToPDBFile(peptide, "peptide.pdb")

# ---- Visualización interactiva con py3Dmol ----
# Cargar la estructura y mostrar en 3D
view = py3Dmol.view(width=600, height=400)
view.addModel(Chem.MolToMolBlock(peptide), 'mol')
view.setStyle({'stick': {}, 'sphere': {'scale': 0.3}})
view.setBackgroundColor('0xeeeeee')
view.zoomTo()
view.show()
```
