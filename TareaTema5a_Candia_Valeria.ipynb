```python
import pubchempy as pcp
from rdkit import Chem
from rdkit.Chem import rdMolDescriptors
import re

def buscar_compuesto(nombre_compuesto):
    """Busca un compuesto en PubChem por nombre o f√≥rmula"""
    try:
        # Primero intenta por nombre
        compuestos = pcp.get_compounds(nombre_compuesto, 'name')
        if not compuestos:
            # Si no encuentra por nombre, intenta por f√≥rmula
            compuestos = pcp.get_compounds(nombre_compuesto, 'formula')
        
        if not compuestos:
            return None
            
        return compuestos[0]
    
    except Exception:
        return None

def obtener_masa_molar(compuesto_pubchem):
    """Calcula la masa molar usando RDKit o PubChem"""
    try:
        smiles = compuesto_pubchem.isomeric_smiles or compuesto_pubchem.canonical_smiles
        
        if smiles:
            mol = Chem.MolFromSmiles(smiles)
            if mol is not None:
                return rdMolDescriptors.CalcExactMolWt(mol)
        
        # Si RDKit falla, usar masa de PubChem
        return compuesto_pubchem.molecular_weight
    
    except Exception:
        return compuesto_pubchem.molecular_weight

def contar_atomos_en_formula(formula, elemento):
    """Cuenta cu√°ntos √°tomos de un elemento hay en una f√≥rmula molecular"""
    # Buscar el patr√≥n del elemento seguido de un n√∫mero (opcional)
    patron = rf'{elemento}(\d*)'
    coincidencias = re.findall(patron, formula)
    
    if not coincidencias:
        return 0
    
    # Si no hay n√∫mero despu√©s del elemento, es 1
    # Si hay n√∫mero, usar ese n√∫mero
    total = 0
    for coincidencia in coincidencias:
        if coincidencia == '':
            total += 1
        else:
            total += int(coincidencia)
    
    return total

def calculomasacompuesto():
    print("========================================")
    print("CALCULADORA DE MASA DE COMPONENTES")
    print("========================================")
    
    # Entrada de datos
    compA = input("Ingrese el componente A (elemento o compuesto): ").strip()
    compB = input("Ingrese el componente B (compuesto que contiene A): ").strip()
    masa_compB = float(input(f"Ingrese la masa de {compB} (g): "))
    
    try:
        # Buscar ambos compuestos
        print(f"\nBuscando informaci√≥n de {compA}...")
        compuesto_A = buscar_compuesto(compA)
        if not compuesto_A:
            print(f"No se encontr√≥ {compA} en PubChem")
            return
        
        print(f"Buscando informaci√≥n de {compB}...")
        compuesto_B = buscar_compuesto(compB)
        if not compuesto_B:
            print(f"No se encontr√≥ {compB} en PubChem")
            return
        
        # Obtener masas molares autom√°ticamente
        masa_molar_compA = obtener_masa_molar(compuesto_A)
        masa_molar_compB = obtener_masa_molar(compuesto_B)
        
        # Determinar coeficiente estequiom√©trico autom√°ticamente
        formula_A = compuesto_A.molecular_formula
        formula_B = compuesto_B.molecular_formula
        
        # Si A es un elemento simple, contar cu√°ntos √°tomos hay en B
        if len(formula_A) <= 3 and formula_A.isalpha():  # Elemento simple como H, C, N, etc.
            coeficiente = contar_atomos_en_formula(formula_B, formula_A)
        else:
            # Si A es un compuesto, buscar cu√°ntas veces aparece en B
            # Para simplificar, asumimos coeficiente 1 (puede mejorarse)
            coeficiente = 1
            
        if coeficiente == 0:
            print(f"Error: {compA} no est√° presente en {compB}")
            return
        
        # C√°lculos estequiom√©tricos
        moles_compB = masa_compB / masa_molar_compB
        moles_compA = moles_compB * coeficiente
        masa_compA = moles_compA * masa_molar_compA
        porcentaje = (masa_compA / masa_compB) * 100
        
        # Mostrar resultados
        print("\n========================================")
        print(f"INFORMACI√ìN ENCONTRADA")
        print("========================================")
        print(f"{compA}: {compuesto_A.molecular_formula}")
        print(f"Masa molar de {compA}: {masa_molar_compA:.2f} g/mol")
        print(f"{compB}: {compuesto_B.molecular_formula}")
        print(f"Masa molar de {compB}: {masa_molar_compB:.2f} g/mol")
        print(f"Coeficiente estequiom√©trico: {coeficiente}")
        
        print("\n========================================")
        print(f"RESULTADOS")
        print("========================================")
        print(f"Moles de {compB}: {moles_compB:.6f} mol")
        print(f"Moles de {compA}: {moles_compA:.6f} mol")
        print(f"Masa de {compA} en {masa_compB}g de {compB}: {masa_compA:.4f} g")
        print(f"Porcentaje en masa: {porcentaje:.2f}%")
        
    except ValueError:
        print("Error: Ingrese un n√∫mero v√°lido para la masa")
    except Exception as e:
        print(f"Error: {str(e)}")

# Ejecutar la funci√≥n
calculomasacompuesto()

```

```python
import pubchempy as pcp
from rdkit import Chem
from rdkit.Chem import rdMolDescriptors
import re

class CalculadoraSimple:
    def __init__(self):
        # Combinaciones predefinidas de compuesto y elemento
        self.combinaciones = {
            1: {"compuesto": "Na2SO4", "elemento": "Na", "descripcion": "Sodio en Sulfato de Sodio"},
            2: {"compuesto": "Na2SO4", "elemento": "S", "descripcion": "Azufre en Sulfato de Sodio"},
            3: {"compuesto": "CaCl2", "elemento": "Ca", "descripcion": "Calcio en Cloruro de Calcio"},
            4: {"compuesto": "CaCl2", "elemento": "Cl", "descripcion": "Cloro en Cloruro de Calcio"},
            5: {"compuesto": "MgSO4", "elemento": "Mg", "descripcion": "Magnesio en Sulfato de Magnesio"},
            6: {"compuesto": "MgSO4", "elemento": "S", "descripcion": "Azufre en Sulfato de Magnesio"},
            7: {"compuesto": "NaCl", "elemento": "Na", "descripcion": "Sodio en Cloruro de Sodio"},
            8: {"compuesto": "NaCl", "elemento": "Cl", "descripcion": "Cloro en Cloruro de Sodio"},
            9: {"compuesto": "KNO3", "elemento": "K", "descripcion": "Potasio en Nitrato de Potasio"},
            10: {"compuesto": "KNO3", "elemento": "N", "descripcion": "Nitr√≥geno en Nitrato de Potasio"},
            11: {"compuesto": "CaCO3", "elemento": "Ca", "descripcion": "Calcio en Carbonato de Calcio"},
            12: {"compuesto": "CaCO3", "elemento": "C", "descripcion": "Carbono en Carbonato de Calcio"},
            13: {"compuesto": "Al2O3", "elemento": "Al", "descripcion": "Aluminio en √ìxido de Aluminio"},
            14: {"compuesto": "Al2O3", "elemento": "O", "descripcion": "Ox√≠geno en √ìxido de Aluminio"},
            15: {"compuesto": "FeSO4", "elemento": "Fe", "descripcion": "Hierro en Sulfato de Hierro"},
            16: {"compuesto": "CuSO4", "elemento": "Cu", "descripcion": "Cobre en Sulfato de Cobre"},
            17: {"compuesto": "NH4Cl", "elemento": "N", "descripcion": "Nitr√≥geno en Cloruro de Amonio"},
            18: {"compuesto": "NH4Cl", "elemento": "Cl", "descripcion": "Cloro en Cloruro de Amonio"},
            19: {"compuesto": "H2SO4", "elemento": "S", "descripcion": "Azufre en √Åcido Sulf√∫rico"},
            20: {"compuesto": "NaOH", "elemento": "Na", "descripcion": "Sodio en Hidr√≥xido de Sodio"}
        }
        
        self.masas_atomicas = {
            'H': 1.008, 'He': 4.003, 'Li': 6.941, 'Be': 9.012, 'B': 10.811,
            'C': 12.011, 'N': 14.007, 'O': 15.999, 'F': 18.998, 'Ne': 20.180,
            'Na': 22.990, 'Mg': 24.305, 'Al': 26.982, 'Si': 28.086, 'P': 30.974,
            'S': 32.065, 'Cl': 35.453, 'Ar': 39.948, 'K': 39.098, 'Ca': 40.078,
            'Fe': 55.845, 'Cu': 63.546, 'Zn': 65.409, 'Br': 79.904, 'I': 126.904,
            'Mn': 54.938, 'Cr': 51.996, 'Ni': 58.693, 'Co': 58.933, 'Pb': 207.2
        }

    def mostrar_menu(self):
        """Muestra el men√∫ de combinaciones disponibles"""
        print("\n" + "="*70)
        print("           CALCULADORA DE MASA - SELECCIONE UNA OPCI√ìN")
        print("="*70)
        
        for i in range(1, len(self.combinaciones) + 1):
            combo = self.combinaciones[i]
            print(f"{i:2d}. {combo['descripcion']} ({combo['compuesto']} ‚Üí {combo['elemento']})")
        
        print("="*70)

    def obtener_elementos_en_compuesto(self, formula):
        """Extrae elementos y sus cantidades de una f√≥rmula qu√≠mica"""
        patron = r'([A-Z][a-z]?)(\d*)'
        coincidencias = re.findall(patron, formula)
        
        elementos = {}
        for elemento, cantidad in coincidencias:
            if cantidad == '':
                cantidad = 1
            else:
                cantidad = int(cantidad)
            
            if elemento in elementos:
                elementos[elemento] += cantidad
            else:
                elementos[elemento] = cantidad
        
        return elementos

    def calcular_masa(self, compuesto, elemento, masa_compuesto):
        """Realiza el c√°lculo completo usando PubChem y RDKit"""
        try:
            # Obtener informaci√≥n del compuesto usando PubChem
            compuesto_info = pcp.get_compounds(compuesto, 'formula')
            if not compuesto_info:
                raise ValueError(f"No se encontr√≥ el compuesto {compuesto}")
            
            # Crear mol√©cula RDKit
            mol = Chem.MolFromSmiles(compuesto_info[0].isomeric_smiles)
            if mol is None:
                raise ValueError(f"No se pudo procesar {compuesto}")
            
            # Calcular masa molecular exacta
            masa_molecular = rdMolDescriptors.CalcExactMolWt(mol)
            
            # Calcular moles del compuesto
            moles_compuesto = masa_compuesto / masa_molecular
            
            # Obtener cantidad del elemento en el compuesto
            elementos = self.obtener_elementos_en_compuesto(compuesto)
            cantidad_elemento = elementos.get(elemento, 0)
            
            if cantidad_elemento == 0:
                raise ValueError(f"El elemento {elemento} no est√° en {compuesto}")
            
            # Calcular moles del elemento
            moles_elemento = moles_compuesto * cantidad_elemento
            
            # Obtener masa molecular real del elemento usando PubChem y RDKit
            try:
                elemento_compuesto = pcp.get_compounds(elemento, 'formula')
                if elemento_compuesto:
                    mol_elemento = Chem.MolFromSmiles(elemento_compuesto[0].isomeric_smiles)
                    masa_molecular_elemento = rdMolDescriptors.CalcExactMolWt(mol_elemento)
                else:
                    # Si no se encuentra en PubChem, usar masa at√≥mica del diccionario
                    masa_molecular_elemento = self.masas_atomicas.get(elemento)
                    if masa_molecular_elemento is None:
                        raise ValueError(f"Masa de {elemento} no disponible")
            except:
                # Fallback a masa at√≥mica del diccionario
                masa_molecular_elemento = self.masas_atomicas.get(elemento)
                if masa_molecular_elemento is None:
                    raise ValueError(f"Masa de {elemento} no disponible")
            
            # Calcular masa del elemento
            masa_elemento = moles_elemento * masa_molecular_elemento
            
            return {
                'masa_elemento': masa_elemento,
                'masa_molecular': masa_molecular,
                'moles_compuesto': moles_compuesto,
                'cantidad_elemento': cantidad_elemento,
                'moles_elemento': moles_elemento,
                'masa_molecular_elemento': masa_molecular_elemento,
                'porcentaje': (masa_elemento / masa_compuesto) * 100
            }
            
        except Exception as e:
            print(f"‚ùå Error: {e}")
            return None

    def mostrar_resultado(self, compuesto, elemento, masa_compuesto, resultado):
        """Muestra el resultado del c√°lculo"""
        print("\n" + "="*60)
        print("                     RESULTADO")
        print("="*60)
        print(f"Compuesto: {compuesto}")
        print(f"Elemento buscado: {elemento}")
        print(f"Masa inicial: {masa_compuesto:.2f} g")
        print("\n" + "üéØ RESPUESTA:")
        print(f"   Masa de {elemento}: {resultado['masa_elemento']:.4f} g")
        print(f"   Porcentaje: {resultado['porcentaje']:.2f}%")
        
        print("\n" + "üìä DETALLES:")
        print(f"   ‚Ä¢ Masa molecular de {compuesto}: {resultado['masa_molecular']:.4f} g/mol")
        print(f"   ‚Ä¢ Moles de {compuesto}: {resultado['moles_compuesto']:.6f} mol")
        print(f"   ‚Ä¢ √Åtomos de {elemento} por mol√©cula: {resultado['cantidad_elemento']}")
        print(f"   ‚Ä¢ Moles de {elemento}: {resultado['moles_elemento']:.6f} mol")
        print(f"   ‚Ä¢ Masa at√≥mica de {elemento}: {resultado['masa_molecular_elemento']:.4f} g/mol")
        print(f"   ‚Ä¢ MASA FINAL DE {elemento}: {resultado['masa_elemento']:.4f} g")
        print("="*60)

    def ejecutar(self):
        """Ejecuta el programa principal"""
        while True:
            try:
                self.mostrar_menu()
                
                # Obtener selecci√≥n del usuario
                opcion = int(input(f"\nSeleccione una opci√≥n (1-{len(self.combinaciones)}): "))
                
                if opcion < 1 or opcion > len(self.combinaciones):
                    print("‚ùå Opci√≥n no v√°lida. Intente de nuevo.")
                    input("Presione Enter para continuar...")
                    continue
                
                # Obtener datos de la combinaci√≥n seleccionada
                combo = self.combinaciones[opcion]
                compuesto = combo['compuesto']
                elemento = combo['elemento']
                descripcion = combo['descripcion']
                
                print(f"\n‚úÖ Seleccionado: {descripcion}")
                
                # Obtener masa del usuario
                masa = float(input(f"Ingrese la masa de {compuesto} en gramos: "))
                
                if masa <= 0:
                    print("‚ùå La masa debe ser positiva.")
                    input("Presione Enter para continuar...")
                    continue
                
                print("\nüîÑ Calculando...")
                
                # Realizar c√°lculo
                resultado = self.calcular_masa(compuesto, elemento, masa)
                
                if resultado:
                    self.mostrar_resultado(compuesto, elemento, masa, resultado)
                
                # Preguntar si continuar
                continuar = input("\n¬øDesea hacer otro c√°lculo? (s/n): ").lower().strip()
                if continuar != 's' and continuar != 'si':
                    print("\n¬°Gracias por usar la calculadora! üëã")
                    break
                
            except ValueError as e:
                print("‚ùå Por favor ingrese un n√∫mero v√°lido.")
                input("Presione Enter para continuar...")
            except KeyboardInterrupt:
                print("\n\n¬°Hasta luego! üëã")
                break
            except Exception as e:
                print(f"‚ùå Error inesperado: {e}")
                input("Presione Enter para continuar...")

# Ejecutar el programa
if __name__ == "__main__":
    calculadora = CalculadoraSimple()
    calculadora.ejecutar()
```
